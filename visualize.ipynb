```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the scraped CSV data
df = pd.read_csv("ipl_highest_totals.csv")

# Extract only the runs from the 'Score' column (ignore wickets)
df["Score"] = df["Score"].str.split("/").str[0]  # Get only the runs
df["Score"] = df["Score"].astype(int)  # Convert to integer

# Sort data in descending order
df = df.sort_values(by="Score", ascending=False)

# Set figure size
plt.figure(figsize=(12, 6))

# Create a line plot
sns.lineplot(x=df["Team"], y=df["Score"], marker="o", linestyle="-", color="b")

# Customize plot
plt.title("Highest Team Totals in IPL", fontsize=14)
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Highest Score", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.show()
```

```python
# Load the scraped CSV data
df = pd.read_csv("ipl_highest_totals.csv")

# Extract only the runs from the 'Score' column (ignore wickets)
df["Score"] = df["Score"].str.split("/").str[0]  # Get only the runs
df["Score"] = df["Score"].astype(int)  # Convert to integer

# Sort data in descending order
df = df.sort_values(by="Score", ascending=False)

# Set figure size
plt.figure(figsize=(12, 6))

# Create barplot
sns.barplot(x=df["Team"], y=df["Score"], palette="coolwarm")

# Customize plot
plt.title("Highest Team Totals in IPL", fontsize=14)
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Highest Score", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.show()
```

```python
# Load the scraped CSV data
df = pd.read_csv("cricket_results_2021.csv")

# Count number of wins for each team
winner_counts = df["Winner"].value_counts()

# Set figure size
plt.figure(figsize=(12, 6))

# Create a bar chart
sns.barplot(x=winner_counts.index, y=winner_counts.values, palette="viridis")

# Customize plot
plt.title("IPL 2021 - Number of Wins Per Team", fontsize=14)
plt.xlabel("Teams", fontsize=12)
plt.ylabel("Number of Wins", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.show()
```

```python
winner_counts = df["Winner"].value_counts()

# Set figure size
plt.figure(figsize=(10, 6))

# Generate distinct colors using seaborn
colors = sns.color_palette("tab10", len(winner_counts))  # Use 'tab10' for distinct colors

# Create a pie chart
plt.pie(winner_counts, labels=winner_counts.index, autopct="%1.1f%%", startangle=140, 
        colors=colors, wedgeprops={"edgecolor": "black"})

# Customize plot
plt.title("IPL 2021 - Win Percentage by Team", fontsize=14)

# Show plot
plt.show()
```
